```python
finaldata = []

web = "https://www.goodreads.com/list/show/10334._Best_Top_Romance_Novels_of_All_Time?page=%s" 
for page in range(1,3):
    url = web % page
    data = html.fromstring(requests.get(url).text)
    
    for i in data.xpath("//tr/td[3]"):
        book = i.xpath('a[@class="bookTitle"]/span[1]/text()')
        bookurl = i.xpath('a[@class="bookTitle"]/@href')
        rating = i.xpath('span/span[@class="minirating"]/text()')
        score = i.xpath('div[1]/span[1]/a[1]/text()')
        finaldata.append([book, bookurl, rating, score])

len(finaldata)
```

still, the above xpath is not perfect: 

- Some bold texts inside **bold tag** are not captured above.

```python
for i in data.xpath("//div[@class='left bodycol']"):
    if i.xpath('div[2]/span[1]/span[2]'):
        rev = i.xpath('div[2]/span[1]/span[2]')
        for a in rev:
            print a.text_content() 
    else:
        rev = i.xpath('div[2]/span[1]/span[1]')
        for a in rev:
            print a.text_content() 
    print "++++++++++++++++++++++"
```

```python
review = []
for i in data.xpath("//div[@class='left bodycol']"):
    if i.xpath('div[2]/span[1]/span[2]'):
        rev = i.xpath('div[2]/span[1]/span[2]')
        for a in rev:
            review.append(a.text_content()) 
    else:
        rev = i.xpath('div[2]/span[1]/span[1]')
        for a in rev:
            review.append(a.text_content())

len(review)
```

Noted that Pride and Prejudice (https://www.goodreads.com/book/show/1885.Pride_and_Prejudice) has hundreds (if not thousands) of customer reviews across several web pages. This demo focuses on customer reviews on the first web page only.

The website uses **java script (advanced web technology, ajax)** so, in order to collect all the customer reviews, you need to use **scrapy**. That is a separate topic to discuss at a later time :) ...

```python
review = []

# let's crawl the first three urls only

for url in urls[:3]:
    r = requests.get(url)
    data = html.fromstring(r.text)
    
    for i in data.xpath("//div[@class='left bodycol']"):
        if i.xpath('div[2]/span[1]/span[2]'):
            rev = i.xpath('div[2]/span[1]/span[2]')
            for a in rev:
                review.append(a.text_content()) 
        else:
            rev = i.xpath('div[2]/span[1]/span[1]')
            for a in rev:
                review.append(a.text_content())

len(review)
```

```python
for i in review:
    print i
```
