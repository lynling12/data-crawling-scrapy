## Web Scraping (first webpage only) 

```python
import requests
from lxml import html
import csv
import pandas as pd
```

```python
r = requests.get('https://www.etsy.com/c/accessories/hair_accessories/fascinators_and_mini_hats?explicit=1&page=1')
data = html.fromstring(r.text)
data
```

```python
all=[]
for i in data.xpath("//div[@class='v2-listing-card__info']"):
    title = i.xpath('p[1]/text()')
    if i.xpath('p[2]/span[2]/span[1]'):
        price = i.xpath('p[2]/span[1]/span[2]/text()')
    else:
        price = i.xpath('p[2]/span[2]/text()')
    print title, price
    all.append([title,price])
```

```python
pd.set_option('display.max_colwidth', -1)
df = pd.DataFrame(all)
df
```

```python
all=[]

web = 'https://www.etsy.com/c/accessories/hair_accessories/fascinators_and_mini_hats?explicit=1&page=%s'

for page in range(1,4):
    url = web % page
    r = requests.get(url)
    data = html.fromstring(r.text)
    #xpath
    for i in data.xpath("//div[@class='v2-listing-card__info']"):
        title = i.xpath('p[1]/text()')
        if i.xpath('p[2]/span[2]/span[1]'):
            price = i.xpath('p[2]/span[1]/span[2]/text()')
        else:
            price = i.xpath('p[2]/span[2]/text()')
        print title, price
        all.append([title,price])
    
len(all)
```

```python
pd.set_option('display.max_colwidth', -1)
df = pd.DataFrame(all)
df
```
